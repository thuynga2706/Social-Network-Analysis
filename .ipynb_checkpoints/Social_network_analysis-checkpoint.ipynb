```python
print(f"all tweets {tweets.shape[0]}" )

#get tweets that are retweeted
# tweets1 = tweets.drop_duplicates(['id'],keep='first')
# print(f"drop duplicates {tweets1.shape[0]}" )


tweets1 = tweets.dropna(subset=['quote_url'])
retweetdf = tweets1[tweets1.quote_url.str.contains('https://twitter')]
print( f'only reweets {retweetdf.shape[0]}')


#get original tweeter
retweetdf['tweetowner']=retweetdf.quote_url.apply(lambda string: re.search('https://twitter.com/(.+?)/status/', string).group(1))


#remove tweets that are retweeted by owner
retweetdf = retweetdf[retweetdf.tweetowner != retweetdf.username]
print(f'remove self reweets {retweetdf.shape[0]}')
```

# Network Analysis

## Monopartite Graph: Users connected to each other

```python
G = nx.DiGraph()

#weighted graph, weight is the number of reweets
relationships = pd.DataFrame(retweetdf[['tweetowner','username']].groupby(['tweetowner','username']).size()).reset_index()
#relationships_1 = retweetdf[['tweetowner','username']].values#.tolist()

G.add_weighted_edges_from(relationships.values)
```

```python
retweetdf.shape,G.number_of_edges()
```

```python
#G.edges(data=True) #(tweetowner,username, weight)
```

```python
plt.figure(figsize=(10,9))
nx.draw(G)
```

```python
#Find Influencers
#Degree Centrality (degree of node / Number of nodes -1)
outdeg = nx.out_degree_centrality(G)
sorted(outdeg.items(),key = operator.itemgetter(1),reverse=True)[0:10]
```

## Bipartite Graph: User and Topics

```python
from networkx.algorithms import bipartite
```

```python
relationshipsb1 = pd.DataFrame(retweetdf.groupby(['tweetowner','search']).count().id).reset_index() # number of reweets
#relationshipsb1 = relationshipsb.sort_values('id',ascending = False)#[0:300]#.tweetowner.to_list()
```

```python
relationshipsb1['tweetowner'] = relationshipsb1['tweetowner'].astype(str).str.upper()
```

```python
relationshipsb1 #id is the number of retweets in each topic
```

```python
#relationshipsb1.groupby('tweetowner').count().search.sort_values().tail(50).index.to_list()
#relationshipsb1 = relationshipsb1[relationshipsb1.tweetowner.isin(relationshipsb1.groupby('tweetowner').count().search.sort_values().tail(50).index.to_list())]
```

```python
B=nx.Graph()
B.add_nodes_from(relationshipsb1.tweetowner.to_list(),bipartite = 0)
B.add_nodes_from(relationshipsb1.search.to_list(),bipartite = 1)

B.add_weighted_edges_from(relationshipsb1.values)
```

```python
#Bipartite Graph, left side is Users, right side is Topics (ugly I know :/)
plt.figure(figsize=(10,9))
nx.draw_networkx(
    B,
    pos = nx.drawing.layout.bipartite_layout(B, relationshipsb1.tweetowner.to_list()))
```

```python
#Find Influencers
degX,degY= bipartite.degrees(B,relationshipsb1.search.to_list(),weight='weight')
sorted(dict(degX).items(),key = operator.itemgetter(1),reverse=True)
```

```python
#Which topics are Reuter famous for? 
B.edges('REUTERS',data=True)
```

```python
#Projected Graph on Topics: Which topics are more likely to come from the same Users?
X = set(relationshipsb1.search)
P=bipartite.projected_graph(B,X,multigraph=True)
Edge = P.edges(data=True)
# for a, b, data in sorted(P.edges(data=True), key=lambda x: x[2]['weight'],reverse = True):#[0:5]:
#     print('{a} {b} {w}'.format(a=a, b=b, w=data['weight']))
```

```python
plt.figure() #which topics are more likely connected?
nx.draw_kamada_kawai(P,with_labels=True,multigraph=True,node_size =90,font_size =10,width=0.1)
```
