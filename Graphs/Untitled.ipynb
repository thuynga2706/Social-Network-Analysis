```python
# Instantiate the graph
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7),
                   (5, 8),
                   (8, 9)])

# draw the network G1
nx.draw_networkx(G1)
```

```python
G = nx.karate_club_graph()
```

```python
G = nx.convert_node_labels_to_integers(G,first_label= 1)
```

```python
G.edges()
```

```python
nx.draw_networkx(G)
```

```python
T = nx.bfs_tree(G,1)
T.edges()
```

```python
#Average Distance - Trung binh qua 1 ban chung la biet nhau
nx.average_shortest_path_length(G)
```

```python
#largest distance from node i
nx.eccentricity(G)
```

```python
#max distance in graph from eccentricity
nx.diameter(G) #return 5
# nodes which have distance = 5
nx.periphery(G)
```

```python
#min distance of all max distances from eccentricity
nx.radius(G) #return 3
# nodes which have distance = 3
nx.center(G) # from these people, we can reach to any one
```
